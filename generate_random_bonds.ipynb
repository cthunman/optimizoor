```python
company_list = [
    ("ETP", 10),
    ("LLOYDS", 8),
    ("APPL", 3),
    ("FB", 3),
    ("MSFT", 3),
    ("ABIBB", 7),
    ("ADI", 8),
    ("AMAT", 7),
]
buckets = [1, 5, 6]
```

```python
tickers = [company_list[np.random.randint(0, len(company_list))][0] for _ in range(30)]
```

```python
num_bonds = 30
```

```python
isin_list = np.array([uuid.uuid4() for _ in range(num_bonds)])
companies = [company_list[np.random.randint(0, len(company_list))] for _ in range(num_bonds)]
ticker_list = np.array([x[0] for x in companies])
rating_score_list = np.array([x[1] for x in companies])

rng = np.random.default_rng()
random_list = rng.random(num_bonds).dot(.1)
yield_list = np.array([3.5 + (rating_score_list[i] * .1) + random_list[i] for i in range(num_bonds)])

amount_offered_list = [5 + np.random.randint(0, 16) for _ in range(num_bonds)]
bucket_list = [buckets[np.random.randint(0, len(buckets))] for _ in range(num_bonds)]
```

```python
df = pd.DataFrame({
    "isin": isin_list,
    "ticker": ticker_list,
    "rating_score": rating_score_list,
    "yield": yield_list,
    "amount_offered": amount_offered_list,
    "bucker": bucket_list,
})
```
